```python
import pandas as pd
from pathlib import Path
```

```python
csv_path = Path("../Resources/people_cleansed.csv")
cleansed_df = pd.read_csv(csv_path)
cleansed_df.head()
```

```python
cleansed_df['Salary'].max()
```

```python
# Create the variable bins to define our bounds for when we cut the dataframe.
bins = [0,20000,40000,60000,80000,100000,120000]
print(bins)
```

```python
# Create names for the bins
names = ['unemployed','intern','junior associate','associate', 'junior partner', 'partner']
print(names)
```

```python
cleansed_df['Employment Level'] = pd.cut(cleansed_df['Salary'], bins=bins, labels=names)
```
